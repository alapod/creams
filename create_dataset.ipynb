```python
all_ingredients = '\n'.join(ing_df['ingredients'].values)
```

```python
seps = [',', 'â€¢', '\n', ' . ', '[]', ';', '•', '"', '<br>', 'ingredients', '........................\t', '*', '\t', ' / ', 'ï½¥', ' ? ']
```

```python
def split(txt, seps, pattern):
    default_sep = seps[0]
    for sep in seps[1:]:
        txt = txt.replace(sep, default_sep)
    pattern = re.compile(pattern)
    txt = re.sub(pattern, '', txt)
    return [i.strip().lower() for i in txt.split(default_sep)]
```

```python
a = split(all_ingredients, seps, '\(.*?\)')
```

```python
len(a)
```

```python
b = Counter(a).most_common()[::-1]
```
